```python
#log_path = "data/monomer_chimeras.log"
log_path = "data/runs2_monomers_mutations.log"
#log_path = "data/monomers_mutations_with_isolates.log"

#MONOMER_NAME: [(MonomerId, mutations)...]
monomers_list = {}
with open(log_path) as fr:
    for line in fr:
        if "MonomerId" in line:
            monomerName = line.split()[0]
            monomerId = line.split("MonomerId= ")[-1].split()[0]
            mutationList = line.split('[ ')[-1].split(' ]')[0].split(', ')
            if mutationList[0] == '':
                mutationList = []
            
            for i in range(len(mutationList)):
                mutationList[i] = mutationList[i].split('(')[0]
            
            if monomerName not in monomers_list:
                monomers_list[monomerName] = []
            
            monomers_list[monomerName].append((monomerId, mutationList))
            
print(monomers_list)
```

```python
horInfoPath = "data/cenX_hordecomposition.tsv"
horPos = []
monomers_cnt = 0
cur_line = 0
with open(horInfoPath) as f:
    for line in f:
        cur_line += 1
        if cur_line < 4:
            continue
        parts = line.split('\t')
        if (int(parts[2]) != -1):
            horPos.append((monomers_cnt, monomers_cnt + int(parts[2]) - 1))
            monomers_cnt += int(parts[2])
print(horPos)
    
```

```python
def getHORId(coord):
    for i in range(len(horPos)):
        if (coord >= horPos[i][0]) and (coord <= horPos[i][1]):
            return i
```

```python
HOR_G = []
for i in range(len(horPos)):
    HOR_G.append([])
```

```python
for monoName in monomers_list:
    monoList = monomers_list[monoName]
    for frM_id in range(len(monoList)):
        hor_id = getHORId(int(monoList[frM_id][0]))
        for mut_pos in monoList[frM_id][1]:
            for scM_id in range(frM_id + 1, min(frM_id + 14, len(monoList))):
                if mut_pos in monoList[scM_id][1]:
                    hor_id2 = getHORId(int(monoList[scM_id][0]))
                    if len(HOR_G[hor_id]) == 0 or (HOR_G[hor_id][-1][0] != hor_id2):  
                        HOR_G[hor_id].append((hor_id2, [str(int(monoList[frM_id][0]) - horPos[hor_id][0]) + "(" + mut_pos +")"], 
                                              [str(int(monoList[scM_id][0]) - horPos[hor_id2][0]) + "(" + mut_pos + ")"]))
                    else:
                        HOR_G[hor_id][-1][1].append(str(int(monoList[frM_id][0]) - horPos[hor_id][0]) + "(" + mut_pos +")")
                        HOR_G[hor_id][-1][2].append(str(int(monoList[scM_id][0]) - horPos[hor_id2][0]) + "(" + mut_pos + ")")
             
print(HOR_G)
```

```python
from graphviz import Digraph

dot_str = "graph RUN2 {\n"
for i in range(len(horPos)):
    if len(HOR_G[i]) > 0:
        dot_str += str(i) + ";\n"

for i in range(len(HOR_G)):
    for j in range(len(HOR_G[i])):
        label = ""
        for h in range(len(HOR_G[i][j][1])):
            label += str(HOR_G[i][j][1][h]) + " " + str(HOR_G[i][j][2][h]) + ";\n"
        dot_str += str(i) + " -- " + str(HOR_G[i][j][0]) + " [label=\"" + label + "\"];\n" 
dot_str += "}\n"
print(dot_str)
```
